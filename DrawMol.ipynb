```python
from rdkit import Chem
from rdkit.Chem import Draw

mol = Chem.MolFromSmiles('CC(=C)[C@@H]1[C@@H]2[C@H]([C@]3([C@@]4(CO4)[C@H]5[C@@H]([C@]3([C@H]1C(=O)O2)O)O5)C)O')
mol
```

```python
from rdkit import Chem
from rdkit.Chem import Draw
# Methamphetamine
mol = Chem.MolFromSmiles('CC(CC1=CC=CC=C1)NC')
mol
```
